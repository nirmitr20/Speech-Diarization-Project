```python
device="cpu"
batch_size=16
compute_type="int8"
```

```python
audio_file = "EnglishConvo.wav"
```

```python
model=whisperx.load_model("large-v2", device, compute_type=compute_type)
```

```python
audio=whisperx.load_audio(audio_file)
result=model.transcribe(audio, batch_size=batch_size)
print(result["segments"])
```

```python
from whisperx.diarize import DiarizationPipeline
```

```python
model_a, metadata=whisperx.load_align_model(language_code=result["language"], device=device)

result= whisperx.align(result["segments"],model_a,metadata, audio,device,return_char_alignments=False)

print(result["segments"])
```

```python
diarize_model=DiarizationPipeline(use_auth_token="HuggingFaceToken", device=device)
```

```python
diarize_segments = diarize_model(audio)
```

```python
diarize_segments
```

```python
import json
import pandas as pd

def interval_overlap(a0, a1, b0, b1):
    left = max(a0, b0)
    right = min(a1, b1)
    return max(0.0, right - left)

def normalize_speakers_in_order(rows):
    mapping = {}
    next_id = 1
    for r in rows:
        spk = r["speaker"]
        if spk not in mapping:
            mapping[spk] = f"SPEAKER_{next_id:02d}"
            next_id += 1
        r["speaker"] = mapping[spk]
    return rows

words = []
if any(("words" in seg and seg["words"]) for seg in result.get("segments", [])):
    for seg in result["segments"]:
        for w in seg.get("words", []):
            if "start" in w and "end" in w and "word" in w:
                words.append({
                    "start": float(w["start"]),
                    "end": float(w["end"]),
                    "text": w["word"].strip()
                })
else:
    for seg in result.get("segments", []):
        if "start" in seg and "end" in seg and seg.get("text", "").strip():
            words.append({
                "start": float(seg["start"]),
                "end": float(seg["end"]),
                "text": seg["text"].strip()
            })

df = diarize_segments.sort_values("start").reset_index(drop=True)
dia_rows = [{
    "idx": i,
    "start": float(df.loc[i, "start"]),
    "end": float(df.loc[i, "end"]),
    "speaker": str(df.loc[i, "speaker"]),
} for i in range(len(df))]

assigned = {i: [] for i in range(len(dia_rows))}
min_word_overlap = 0.05  # seconds

for w in words:
    w_s, w_e = w["start"], w["end"]
    if w_e <= w_s:
        continue

    best_i, best_ov = None, 0.0
    for i, row in enumerate(dia_rows):
        ov = interval_overlap(w_s, w_e, row["start"], row["end"])
        if ov > best_ov:
            best_ov = ov
            best_i = i

    if best_i is not None and best_ov >= min_word_overlap:
        assigned[best_i].append(w)

for i in range(len(dia_rows)):
    assigned[i].sort(key=lambda x: (x["start"], x["end"]))

duration_asr = max((w["end"] for w in words), default=0.0)
duration_dia = max((r["end"] for r in dia_rows), default=0.0)
duration = round(max(duration_asr, duration_dia), 2)

segments_out = []
for i, row in enumerate(dia_rows):
    text = " ".join(w["text"] for w in assigned[i]).strip()
    text = " ".join(text.split())

    segments_out.append({
        "start": round(row["start"], 2),
        "end": round(row["end"], 2),
        "speaker": row["speaker"],
        "role": "unknown",   # placeholder
        "text": text
    })

segments_out = normalize_speakers_in_order(segments_out)

role_map = {
    "SPEAKER_01": "victim",
    "SPEAKER_02": "scammer",
    "SPEAKER_03": "DEA",
}
for seg in segments_out:
    seg["role"] = role_map.get(seg["speaker"], "unknown")

result_json = {
    "file_id": "conv_001",
    "duration": duration,
    "segments": segments_out
}

print(json.dumps(result_json, indent=2, ensure_ascii=False))
```

```python
# Save
output_path = "speakers3_English.json"

with open(output_path, "w", encoding="utf-8") as f:
    json.dump(result_json, f, indent=2, ensure_ascii=False)

print(f"Transcript saved successfully to {output_path}")
```

```python
import torch, whisperx, json

audio_path = "Fraud.wav"

device = "cuda" if torch.cuda.is_available() else "cpu"
compute_type = "float16" if device == "cuda" else "int8_float32"  # CPU cannot do float16 efficiently

# 1) Collabora's Hindi fine-tuned model (CTranslate2 format)
model = whisperx.load_model(
    "collabora/faster-whisper-large-v2-hindi",   # also: small/base/medium variants exist
    device=device,
    compute_type=compute_type
)
```

```python
# 2) Transcribe (Hindi)
result = model.transcribe(audio_path, language="hi")

result

# 3) Word-level alignment
model_a, metadata = whisperx.load_align_model(language_code="hi", device=device)
result_aligned = whisperx.align(result["segments"], model_a, metadata, audio_path, device=device)
```

```python
result
```

```python
# 4) Speaker diarization
diarize_model=DiarizationPipeline(use_auth_token="HuggingFaceToken", device=device)
diarize_segments = diarize_model(audio_path)

# 5) Merge speakers with words
result_diarized = whisperx.assign_word_speakers(diarize_segments, result_aligned)
```
